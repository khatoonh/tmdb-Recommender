```python
import ast
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
type(df.iloc[0].crew)
```

```python
# to convert the string to list for using that function
#
#ast.literal_eval()
```

```python
df['genres']=df['genres'].apply(convert)
```

```python
df.head()
```

```python
df['keywords']=df['keywords'].apply(convert)
```

```python
# To extract the only 3 names from cast
import ast
def convert_cast(obj):
    L=[]
    c=0
    for i in ast.literal_eval(obj):
        if c!=3:
            L.append(i['name'])
            c+=1
        else:
            break
    return L
```

```python
df['cast']=df['cast'].apply(convert_cast)
```

```python
import ast
def get_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L
```

```python
df['crew']=df['crew'].apply(get_director)
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
wnl=WordNetLemmatizer()
```

```python
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
def lem(text):
    x=[]
    for i in text.split():
        x.append(wnl.lemmatize(i))
    return " ".join(x)
```

```python
new_df['tags'].apply(stem)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```
